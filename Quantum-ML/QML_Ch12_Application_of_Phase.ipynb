### Basic Application of Amplitude (1 Qubit)

```python
from qiskit import QuantumCircuit, Aer, execute
from qiskit.visualization import plot_histogram, plot_bloch_vector, plot_state_qsphere
import matplotlib.pyplot as plt
qc=QuantumCircuit(1)
qc.h(0)

results=execute(qc, Aer.get_backend('statevector_simulator')).result().get_counts()
plot_histogram(results)
```

```python
plot_state_qsphere(qc)
```

```python
qc=QuantumCircuit(1)
qc.h(0)
qc.z(0)
results=execute(qc, Aer.get_backend('statevector_simulator')).result().get_counts()
plot_state_qsphere(qc)
```

### Oracle that finds |10>

```python
# Two‐qubit Grover searching |10>
from qiskit import QuantumCircuit, Aer, execute
from qiskit.visualization import plot_histogram, plot_bloch_vector, plot_state_qsphere
import matplotlib.pyplot as plt

qc=QuantumCircuit(2)
qc.h(0)
qc.h(1)
qc.barrier()

qc.x(0)
qc.cz(0,1)
qc.x(0)

qc.barrier()
qc.h(0)
qc.h(1)

qc.x(0)
qc.x(1)
qc.cz(0,1)
qc.x(0)
qc.x(1)

qc.barrier()
qc.h(0)
qc.h(1)

results=execute(qc, Aer.get_backend('statevector_simulator')).result().get_counts()
plot_histogram(results)
```

```python
print("Oracle that finds |10>")
qc.draw('mpl')
```

```python
plot_state_qsphere(qc)
```
